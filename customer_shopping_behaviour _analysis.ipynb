```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
#replACE ALL THE NaN VALUES WITH MEDIAN instead of mean values to avoid bias.
```

```python
df.isnull().sum() 
#check if still null values are present
```

```python
df.columns=df.columns.str.lower() #convert everything in lower case (names of the columns)
df.columns=df.columns.str.replace(' ','_') #repLace all the spaces with underscores
```

```python
df.columns  #display title of all the columns
```

```python
#FEATURE ENGINEERING
#CREATE A COLUMN age_group that splits "age" column into a group of four and  furthur classifies them as per the labels that we will provide
labels=["young adult","adult","middle aged", "senior"]
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)   #display the specified columns
```

```python
# create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
